```python
sales_threshold1 = 5
sales_threshold2 = 10
sales_threshold3= 15
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df['Sales_Category'] = pd.cut(df['Sales'], bins=[-float('inf'), sales_threshold1, sales_threshold2, sales_threshold3, float('inf')], labels=['Low', 'Medium', 'High','Highest'])
```

```python
label_encoder = LabelEncoder()
df['ShelfLoc'] = label_encoder.fit_transform(df['ShelveLoc'])
df['Urban'] = label_encoder.fit_transform(df['Urban'])
df['US'] = label_encoder.fit_transform(df['US'])
```

# feature vector and target variable

```python
df3=pd.get_dummies(df, columns=['Urban','ShelveLoc','US'], drop_first=True)
df3
```

```python
X= df3.drop(['Sales_Category'], axis=1)
y= df3['Sales_Category']
```

# Split data into training and test data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test= train_test_split(X,y, test_size= 0.33, random_state =42)
```

# Random model classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
rfc =RandomForestClassifier(random_state=0)
rfc.fit(X_train, y_train)
y_pred= rfc.predict(X_test)

print('Model accuracy score with 10 decision-trees : {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

```python
from sklearn.metrics import confusion_matrix
y_pred = rfc.predict(X_test)
print("Test Accuracy: ",accuracy_score(y_test,y_pred))
print("Test F1 Score: ",f1_score(y_test,y_pred, average='weighted'))
print("Confusion Matrix on Test Data")
q = pd.crosstab(y_test, y_pred, rownames=['True'], colnames=['Predicted'], margins=True)
print(q)
```

```python
df2['Income_Category'] = df2['Taxable.Income'].apply(lambda x: 'Risky' if x <= 30000 else 'Good')
df2 = df2.drop('Taxable.Income', axis=1)

# converts categorical to numerical.
df3 = pd.get_dummies(df2, columns=['Undergrad', 'Marital.Status', 'Urban' ], drop_first=True)

X = df3.drop('Income_Category', axis=1) 
y = df3['Income_Category']
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.tree import DecisionTreeClassifier, export_text
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# decision tree model
classifier_DT = RandomForestClassifier(max_depth=3,random_state= 3)
classifier_DT.fit(X_train, y_train)
y_pred = classifier_DT.predict(X_test)


accuracy = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)
cl = classification_report(y_test, y_pred)
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", cm)
print("\nClassification Report:\n", cl)


```
